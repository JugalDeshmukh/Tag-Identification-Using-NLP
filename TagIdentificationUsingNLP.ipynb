## NLP Overview Theory

### 1. NLP is basically understanding and replying to the human language. NLP is a part of computer science and artificial intelligence that deals with human language.

### 2. Natural Language Understanding (NLU) - NLU is a branch of Artificial intelligence  that uses computer software to understand input made in the form of sentences in text or speech format.

### 3. Natural Language Generation (NLG) - NLG is a software process that transforms structured data into natural language.  It can be used to produce long form content for organizations to automate custom reports as well produce custom content for web or mobile.


# NLP Terminology

### 1. Tokenization - Breaking strings into small individual tokens or words. Eg. Whats the time now -- Whats, the, time , now 
### 2. Stemming - Normalize words into its base form or root form. Eg. Playing, played,plays -- to play
### 3. Lemmatization - Grouping different inflected form of word called Lemma. Similar to stemming but returns the perfect word.  Eg. Better, super -- Good
### 4. POS Tagging (Parts of Speech) - The most popular POS Tagging will be identifying words as nouns , verbs, adjectives etc. Eg. Google about Symbiosis -- Google may be Noun/Verb
### 5. Named Entity Recognition - Recognizing the words as movie,monetory value, organization,location, quantity or person. Eg. Google about Symbiosis -- Google-Verb,Symbiosis - Organization
### 6. Chunking - Picking pieces of words and form into phrases. Whats, the, time , now -- Whats the time now

# Tag Identification Using NLP Project

```python
import urllib.request #Handling URL

from bs4 import BeautifulSoup #Handling or parsing html files

import nltk #toolkit
nltk.download('stopwords') #or is was extracted

from nltk.corpus import stopwords
```

```python
#get the info from website
response =  urllib.request.urlopen('https://en.wikipedia.org/wiki/Microsoft')
html = response.read()
```

```python
## Remove the tags from the website text and get the website textinto a particular text format
soup = BeautifulSoup(html,'html5lib')
text = soup.get_text(strip = True)
```

```python
## Converting the website text into tokens splitting with a comma (,) delimeter
tokens = [t for t in text.split()]
```

```python
sr= stopwords.words('english')

# Remove the stopwords from the website text regardless of the case
clean_tokens = [token for token in tokens if token.lower() not in sr]
```

```python
## Print the frequency of each word
freq = nltk.FreqDist(clean_tokens)
for key,val in freq.items():
    print(str(key) + ':' + str(val))
```

```python
## Plot the frequency plot of the  top 30 tags (words) from the website text
freq.plot(30, cumulative=False)
```
